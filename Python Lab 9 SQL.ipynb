```python
rentals_june = rental[rental["rental_date"].apply(lambda x: "-06-" in x)]
counts_june = rentals_june.groupby('customer_id')['rental_id'].nunique()
final_june = counts_june.reset_index(name='rental_count')
final_june
```

```python
Merged = pd.merge(final_may,final_june, on = 'customer_id')
Merged
```
